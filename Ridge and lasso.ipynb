```python
# Import the required Libraries
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('50_Startups.csv')
```

```python
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['State']=le.fit_transform(df['State'])
```

```python
df.head()
```

```python
x=df.drop(columns=['Profit'])
```

```python
x.head()
```

```python
y=df.Profit
```

```python
y.head()
```

```python
#Normalisation Technique to scale down the values
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
scaled_x=pd.DataFrame(scale.fit_transform(x))
```

```python
scaled_x
```

```python
#Train and Test Split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#importingh Ridge and Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r=Ridge()
l=Lasso()
```

```python
r.fit(x_train,y_train)# Ridge Training
```

```python
l.fit(x_train,y_train) #lasso trainng
```

```python
pred1=r.predict(x_test)
pred2=l.predict(x_test)
```

```python
pred1
```

```python
pred2
```

```python
from sklearn import metrics
```

```python
#MSE (Mean Square Error)
print(metrics.mean_squared_error(y_test,pred1)) # RIDGE
print(metrics.mean_squared_error(y_test,pred2)) #LASSO
```

```python
#RMSE(Root Mean Square Error)
print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```

```python
# R Squared
print(metrics.r2_score(y_test,pred1))# RIDGE
print(metrics.r2_score(y_test,pred2)) #LASSO
```

```python
# As r2 Score is higher for lasso regression we will use Lasso Regression ML Model for Deployment.
```
